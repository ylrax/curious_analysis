```python
import pandas as pd
```

```python
# www.ine.es/daco/daco42/nombyapel/nombres_por_edad_media.xls
df = pd.read_csv("C:\\Users\\3004822\\Desktop\\mujeres.csv", sep=";", encoding="latin")
nom = list(df.nombre)
df.shape
```

```python
def verifier(name, all_list_names):
    """ Return the number of names from the whole list (all_list_names) with letters matches in the name parameter"""
    try:
        tmp = sum([len(list(set(list(name)).intersection(list(str(i))))) != 0 for i in all_list_names])
        return tmp
    except TypeError:
        return "wrong something"
```

```python
df["matches"] = df.nombre.map(lambda x: verifier(x, nom)) #Takes a while...
print("Done")
```

```python
df[df.matches=="wrong something"]
```

```python
df = df[df.matches!="wrong something"]
```

```python
df["lenght"] = df.nombre.map(lambda x: len(x))
df["complex"] = df.nombre.apply(lambda x: ' ' in x)
# df["percent"] = 100*float(df["matches"])/df.shape[0]
```

## Top Vocals
<br><a id="letters"></a>

```python
print("Letter A usage: {:5d} of {:5d})".format(verifier("A", nom),len(nom)))
print("Letter E usage: {:5d} of {:5d})".format(verifier("E", nom),len(nom)))
print("Letter I usage: {:5d} of {:5d})".format(verifier("I", nom),len(nom)))
print("Letter O usage: {:5d} of {:5d})".format(verifier("O", nom),len(nom)))
print("Letter U usage: {:5d} of {:5d})".format(verifier("U", nom),len(nom)))
```

# Restriction to hight frecuentcy names<a id="restricted"></a>
<br>

```python
common_names_df = df[df.Frecuencia>500]
```

```python
common_names = list(common_names_df.nombre)
common_names_df.shape
```

```python
common_names_df["frec_matches"] = common_names_df.nombre.map(lambda x: verifier(x, common_names))
```

# Restricted results

```python
common_names_df.sort_values(by=["frec_matches"], ascending=False).head(10)
```

```python
common_names_df[~common_names_df.complex].sort_values(by=["frec_matches"], ascending=False).head(10)
```

```python
common_names_df[common_names_df.lenght<7].sort_values(by=["frec_matches"], ascending=False).head(15)
```
